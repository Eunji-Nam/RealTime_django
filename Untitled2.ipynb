```python
import os
os.environ["DJANGO_SETTINGS_MODULE"] = "RealTime_django.settings"
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
import django
django.setup()
```

```python
from openpyxl import load_workbook
wb = load_workbook('realtime_review_data.xlsx')
sheet = wb.active
rows = [
   [column.value for column in row]
   for row in sheet
]
rows
```

```python
column_names = rows[0]
column_names
```

```python
row_list = []
for row in rows[1:]:
    if row[0] is None:
        break
    row_dict = dict(zip(column_names, row))
    row_list.append(row_dict)
row_list
```

```python
from shop.models import Review
```

```python
for row_dict in row_list:
    review = Review(**row_dict)
    review.save()
    print(review)
```
